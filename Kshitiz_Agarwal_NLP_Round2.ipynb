```python
import string

# Open the file and read the text
with open('India that is Bharat by J Sai Deepak.txt', 'r') as file:
    text = file.read().replace('\n', ' ')

# Create a string which has all the punctuations to be removed
punctuations = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
T = ""
for char in text:
    if char not in punctuations:
        T = T + char

# Convert the text into lower case
T = T.lower()

# Remove excessive whitespace
T = ' '.join(T.split())
```

```python
# Install spaCy
!pip install -U spacy

# Download the English model for spaCy
!python -m spacy download en_core_web_sm

# Import the necessary libraries
import spacy

# Load the English model
nlp = spacy.load('en_core_web_sm')

# Perform entity recognition
doc = nlp(T)  # Here 'T' is your preprocessed text

# Print the entities detected
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
# Perform entity recognition
doc = nlp(T)  # Here 'T' is your preprocessed text

# Create a set to store unique entity labels
unique_entity_labels = set()

# Iterate over the entities
for ent in doc.ents:
    # Add the entity label to the set
    unique_entity_labels.add(ent.label_)

# Print the unique entity labels
for label in unique_entity_labels:
    print(label)
```
