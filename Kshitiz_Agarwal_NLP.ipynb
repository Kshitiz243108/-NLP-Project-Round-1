```python
import string

# Open the file and read the text
with open('India that is Bharat by J Sai Deepak.txt', 'r') as file:
    text = file.read().replace('\n', ' ')

# Create a string which has all the punctuations to be removed
punctuations = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
T = ""
for char in text:
    if char not in punctuations:
        T = T + char

# Convert the text into lower case
T = T.lower()

# Remove excessive whitespace
T = ' '.join(T.split())



```

```python

import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Assuming 'T' is your text
T = re.sub(r'\d+ [\w\s,]+:', '', T)

# Tokenize the text
tokens_T = word_tokenize(T)

# Remove stop words
stop_words = set(stopwords.words('english'))
tokens_final_T = [i for i in tokens_T if not i in stop_words]

# Join the tokens back into a string
T = ' '.join(tokens_final_T)

```

```python
import matplotlib.pyplot as plt
from nltk.probability import FreqDist

# Assuming 'tokens_final_T' is your list of tokens
fdist = FreqDist(tokens_final_T)

# Create histogram
plt.figure(figsize=(10, 4))
fdist.plot(30)

# Show the plot
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Assuming 'tokens_final_T' is your list of tokens
tokens = ' '.join(tokens_final_T)

# Generate word cloud
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = None,
                min_font_size = 10).generate(tokens)

# Plot the WordCloud image
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

```python
import nltk
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Assuming 'tokens_final_T' is your list of tokens
nltk.download('averaged_perceptron_tagger')
pos_tags = nltk.pos_tag(tokens_final_T)

# Extract tags
tags = [tag for word, tag in pos_tags]

# Create frequency distribution
fdist = FreqDist(tags)

# Print the 10 most common tags
print(fdist.most_common(10))

# Create histogram
plt.figure(figsize=(10, 4))
fdist.plot(30)

# Show the plot
plt.show()
```

```python
import nltk
from nltk import bigrams
from nltk.probability import FreqDist, ConditionalFreqDist
import random

# Import the text
with open('India that is Bharat by J Sai Deepak.txt', 'r') as file:
    T = file.read().replace('\n', ' ')

# Here, you need to extract the specific chapter from T. This depends on how your chapters are separated in the text file.
# For example, if each chapter starts with a specific title, you can do:
start_index = T.index("The discussions undertaken hitherto")
#beginning of chapter 10
# Assuming that the next chapter starts with "2 ", you can find the end index of the current chapter:
end_index = T.index("relation to the GoI Act", start_index)
#end of chapter 10
# Extract the chapter:
C = T[start_index:end_index]

# Tokenise C
nltk.download('punkt')
word_tokens = nltk.word_tokenize(C)

# Create bigrams
bi_grams = list(bigrams(word_tokens))

# Create frequency distribution of bigrams
bi_fdist = FreqDist(bi_grams)

# Create a bi-gram probability table
bi_cfd = ConditionalFreqDist(bi_grams)
for condition in bi_cfd:
    total = sum(bi_cfd[condition].values())
    for sample in bi_cfd[condition]:
        bi_cfd[condition][sample] /= total

# Define a function to generate a sentence for the Shannon game
def generate_sentence(cfdist, word, num=15):
    sentence = []
    for i in range(num):
        sentence.append(word)
        word = cfdist[word].max()

    return ' '.join(sentence)

# Play the Shannon game
start_word = random.choice(list(bi_cfd.keys()))  # Choose a random start word
print(generate_sentence(bi_cfd, start_word))
```
